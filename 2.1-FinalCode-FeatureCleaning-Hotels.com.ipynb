```python
hotelalldf = hoteldf.dropna()
print(hotelalldf.shape)
hotelalldf.sample(5)
```

```python
hotelalldf['price'] = hotelalldf['price'].replace({'\$':''},regex=True)
hotelalldf.head(5)
```

```python
#cleaning hotel star ,grab just numbers with decimal
def remove_chars(val):
    new_val = re.findall('[0-9.]+',val)
    return "".join(new_val)

hotelalldf.hotel_star = hotelalldf.hotel_star.apply(remove_chars)
hotelalldf.head(5)
```

```python
def remove_special_chars(val):
    new_val = re.findall('[0-9]',val)
    return "".join(new_val)
hotelalldf.total_reviews = hotelalldf.total_reviews.apply(remove_special_chars)
hotelalldf.head(5)
```

```python
hotelalldf['rating_cat'] = hotelalldf['rating']
hotelalldf.head()
```

```python
L = ['Very Good','Superb','Exceptional','Fabulous','Good']
pat = '|'.join(r"\b{}\b".format(x) for x in L)
hotelalldf['rating_cat'] = hotelalldf['rating_cat'].str.extract('('+ pat + ')', expand=False)
```

```python
hotelalldf.head()
```

```python
# extract number of rooms
hotelalldf.total_rooms = hotelalldf.total_rooms.str.split(' ').str[0]
hotelalldf.head(5)
```

```python
hotelalldf['occupancy1'] = hotelalldf['occupancy'].replace({'Mon October 26 - Tue October 27, 2020, 1 night, ':''},regex=True)
hotelalldf.head()
```

```python
hotelalldf['checkin'] = hotelalldf['checkin'].replace({'Check-in time ':''},regex=True)
hotelalldf.head()
```

```python
hotelalldf['checkout'] = hotelalldf['checkout'].replace({'Check-out time is ':''},regex=True)
hotelalldf.head()
```

```python
hotelalldf['rating_avg'] = hotelalldf['rating'].str.extract(r'([1-9].[0-9])',expand=False)
hotelalldf.head()
```

```python
hotelalldf.rating_avg = hotelalldf.rating_avg.astype(float)
hotelalldf.hotel_star = hotelalldf.hotel_star.astype(float)
hotelalldf.price = hotelalldf.price.astype(int)
hotelalldf.total_reviews = hotelalldf.total_reviews.astype(int)
hotelalldf.total_rooms = pd.to_numeric(hotelalldf.total_rooms,errors='coerce')
hotelalldf.head(3)
```

```python
def split_to_list(str_val):
    return str_val.split(",")
   
hotelalldf = hotelalldf.assign(occupancy_lst = hotelalldf.occupancy.apply(split_to_list))
```

```python
# just initailzing new cols, assigned values are just place holders
hotelalldf = hotelalldf.assign(no_of_rooms = 1, no_of_guests =  2)
hotelalldf.sample(2)
```
